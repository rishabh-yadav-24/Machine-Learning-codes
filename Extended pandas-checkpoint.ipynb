```python
from seaborn import load_dataset
```

```python
T = load_dataset('titanic')
```

```python
T.head()
```

```python
T.deck.isnull().sum()
```

```python
T.drop(['deck'],axis = 1, inplace=True)
```

```python
T.groupby(['sex','class'])['survived'].mean().unstack()
```

```python
T.pivot_table('survived',index='sex',columns='class',aggfunc='mean')
```

```python
agePartitions = pd.cut(T['age'],[0,18,80])
```

```python
agePartitions
```

```python
T.pivot_table('survived',['sex',agePartitions],'class')
```

```python
T.head()
```

```python
farePartions = pd.qcut(T['fare'],2)
```

```python
T.pivot_table('survived',['sex',agePartitions],[farePartions,'class'])
```

```python
T.pivot_table(index='sex',columns='class',aggfunc={'survived':sum,'fare':'mean'})
```
