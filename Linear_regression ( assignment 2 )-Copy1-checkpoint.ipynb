```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression

X,Y = make_regression(n_samples = 50 , n_features = 1 , n_informative = 1 , n_targets = 1 , noise = 15 , random_state = 13)

print( "shape of X is :" , X.shape)

Y = Y.reshape(X.shape[0],1)
print( "shape of Y is :" ,Y.shape)

aug = np.ones((X.shape[0],1))

X = np.hstack((aug , X))
print("shape of augmented X is : ", X.shape)
```

```python
plt.scatter(X[:,1],Y,color="g")
plt.title('Linear Regression')
plt.xlabel('Numbers of hours studied')
plt.ylabel('Rank of the Student')
plt.plot()
```

```python
print("value of w vector for minumum loss is : ",np.dot((la.inv(np.dot(X.T , X))) , np.dot(X.T , Y)))
```

```python
lr = 0.1
tolerance = 0.0001
count = 0
```

```python
W = np.array([[0],
              [1]])
while True:
    Y_hat = np.dot(X , W)
    L = (np.dot(((Y - Y_hat).T),(Y - Y_hat)))/Y.shape[0]
    partial = -2*(np.dot(((Y - Y_hat).T),X))/Y.shape[0]  
    plt.scatter(X[:,1],Y,color="g")
    plt.plot()
    plt.title('Gradient Descent Method')
    plt.xlabel('Numbers of hours studied')
    plt.ylabel('Rank of the Student')
    p = np.linspace(-3,4,10)
    plt.plot(p,(W[0] + W[1]*p),'-c')
    
    W = W - lr * (partial.T)
    
    
    if np.linalg.norm(lr* partial) < tolerance:
        plt.plot(p,(W[0] + W[1]*p),'-r')
        break
    count = count + 1
    
```

```python
print("Optimal parameter value of W vector is :",W)
print("No of iteration performed :",count)
```

## Now fit a 4th order polynomial function

```python
# f(x;w) = w0 + w1x + w2x^2 + w3x^3 + w4x^4

from sklearn.datasets import make_regression

X,Y = make_regression(n_samples = 50 , n_features = 1 , n_informative = 1 , n_targets = 1 , noise = 15 , random_state = 13)

print( "shape of X is :" , X.shape)

Y = Y.reshape(X.shape[0],1)
print( "shape of Y is :" ,Y.shape)

aug = np.ones((X.shape[0],1))

X = np.hstack((aug , X))
print("shape of augmented X is : ", X.shape)

degree_of_polynomial = 4
X_1col = X[:,1].reshape((50,1))

for i in range(2,degree_of_polynomial+1):
    X_icol = (X_1col)**i
    X = np.hstack((X,X_icol))


print(X.shape,Y.shape)
```

```python
plt.scatter(X[:,1],Y,color="g")
plt.title('Linear Regression')
plt.xlabel('Numbers of hours studied')
plt.ylabel('Rank of the Student')
plt.plot()
```

## Fit high order polynomials

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from numpy import linalg as la
no_of_data = 40
X, Y = make_regression(n_samples=no_of_data, n_features=1, n_informative=1, n_targets=1, noise=15, random_state=13)

Y = Y.reshape(X.shape[0], 1)

aug = np.ones((X.shape[0], 1))
X = np.hstack((aug, X))

X_1col = X[:, 1].reshape((X.shape[0], 1))

for degree_of_polynomial in range(1, 17):
    X_temp = X.copy()
    for i in range(2, degree_of_polynomial + 1):
        X_icol = (X_1col ** i).reshape((X.shape[0], 1))
        X_temp = np.hstack((X_temp, X_icol))

    W = np.dot((la.inv(np.dot(X_temp.T, X_temp))), np.dot(X_temp.T, Y))
    coefficients = W[::-1]
    x_values = np.linspace(-2.5, 2.5, 200)
    y_values = np.polyval(coefficients, x_values)

    plt.scatter(X[:, 1], Y, color="g")
    plt.plot(x_values, y_values, label="Polynomial Funtion")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f"Polynomial Degree: {degree_of_polynomial}")
    plt.grid(True)
    plt.ylim(-200, 200)
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from numpy import linalg as la

X, Y = make_regression(n_samples=50, n_features=1, n_informative=1, n_targets=1, noise=15, random_state=13)

Y = Y.reshape(X.shape[0], 1)

aug = np.ones((X.shape[0], 1))
X = np.hstack((aug, X))

X_1col = X[:, 1].reshape((50, 1))
max_degree = 6
i = 0
training_errors = np.zeros((max_degree))

for degree_of_polynomial in range(1, max_degree):
    X_temp = X.copy()
    for i in range(2, degree_of_polynomial + 1):
        X_icol = (X_1col ** i).reshape((50, 1))
        X_temp = np.hstack((X_temp, X_icol))
        
    
    W = np.dot((la.inv(np.dot(X_temp.T, X_temp))), np.dot(X_temp.T, Y))
    
    Y_hat = np.dot(X_temp,W)
    L = (np.dot(((Y - Y_hat).T),(Y - Y_hat)))/Y.shape[0]
    training_errors[degree_of_polynomial] = L[0][0]
    
    
    coefficients = W[::-1]
    x_values = np.linspace(-2.5, 2.5, 200)
    y_values = np.polyval(coefficients, x_values)

    plt.plot(x_values, y_values, label=f"Degree {degree_of_polynomial}")

plt.scatter(X[:, 1], Y, color="g", label="Data points")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Polynomial Regression with Different Degrees")
plt.ylim(-40, 40)
plt.grid(True)
plt.legend()
plt.show()
```

## Using Cross Validation

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from numpy import linalg as la


no_of_data = 30
X, Y = make_regression(n_samples=no_of_data, n_features=1, n_informative=1, n_targets=1, noise=15, random_state=13)
Y = Y.reshape(X.shape[0], 1)

# Split the data into training and validation sets
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)

aug = np.ones((X_train.shape[0], 1))
X_train = np.hstack((aug, X_train))

aug = np.ones((X_val.shape[0], 1))
X_val   = np.hstack((aug, X_val))

X_1col = X_train[:, 1].reshape((X_train.shape[0], 1))

 = 0.05 # tuning parameter

for degree_of_polynomial in range(1, 17):
    X_temp = X_train.copy()
    for i in range(2, degree_of_polynomial + 1):
        X_icol = (X_1col ** i).reshape((X_train.shape[0], 1))
        X_temp = np.hstack((X_temp, X_icol))
        
    temp =  * np.identity(X_temp.shape[1])
    W = np.dot((la.inv(np.dot(X_temp.T, X_temp) + temp)), np.dot(X_temp.T, Y_train))
    coefficients = W[::-1]
    x_values = np.linspace(-2.5, 2.5, 200)
    y_values = np.polyval(coefficients, x_values)

    plt.scatter(X_train[:, 1], Y_train, color="g", label="Training data")
    plt.scatter(X_val[:, 1], Y_val, color="b", label="Validation data")
    plt.plot(x_values, y_values, label=f"Polynomial Degree: {degree_of_polynomial}")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title("Linear Regression with Increasing Complexity")
    plt.grid(True)
    plt.ylim(-200, 200)
    plt.legend()
    plt.show()
```
